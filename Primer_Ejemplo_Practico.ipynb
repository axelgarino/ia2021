## Numpy


#### Crear un vector con valores de 10 a 49.



```python
vector1 = np.linspace(10,49,40)
vector1
```

#### Elevar al cuadrado a aquellos n√∫meros divisibles por 3.


```python
for i in range(len(vector1)):
    if vector1[i]%3==0:
        vector1[i]=vector1[i]**2
vector1
```

#### Redimensionar ese vector en una matriz de 20x2.


```python
vector1= np.reshape(vector1,(20,2))
vector1
```

#### Crear una nueva matriz de 2x10 y multiplicarla a la matriz anterior.


```python
vector2= np.reshape(np.linspace(1,20,20),(2,10))

vector2
```

```python
vector3 = np.dot(vector1,vector2)
vector3
```

#### Obtener la transpuesta de la matriz resultante.

```python
vector3=np.transpose(vector3)
vector3
```
